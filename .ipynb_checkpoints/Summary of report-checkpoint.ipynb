## Project Summary
### 1) Business Case: 
With more than 15 years of international business experience, "INX Future Inc.", also known as INX, is one of the top providers of automation and data analytics solutions. The health of the employee performance indices has declined recently, which is concerning upper management more and more. Increased service delivery escalations have resulted in an 8 percentage point decline in client satisfaction.The company's CEO made the decision to examine the available employee data in order to identify the fundamental reasons behind the employees' performance problems. In order to prevent any necessary punishment of a non-performing employee from gravely impairing the morale of other employees, the company also anticipates clear symptoms of non-performance. 
##### The Goal and Insights of the project are as follows:

* Department wise performances
* Top 3 Important Factors effecting employee performance
* A trained model which can predict the employee performance based on factors as inputs. This will be used to hire employees
* Recommendations to improve the employee performance based on insights from analysis

### 2)Requirement
Data from third party sources. The employee performance data of INX Future Inc. can be downloads from below link. http://data.iabac.org/exam/p2/data/INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls

### 3) Analysis

 By describing the features found in the data, the data were analysed. In the analysis, the traits are more important. The relationship between the independent and dependent variables is revealed by the characteristics. Moreover, pandas aid in the dataset descriptions that address the following queries at the beginning of our project. The dataset contains two types of data: categorical and numerical.  The target variable 'PerformanceRating' is ordinal.
### Exploratory data analysis
#### Univariant Analysis
* The maximum employee age range of INX Future Inc. is between 25 and 45 years old, with an average employee age of 36.
* Many employee live close to the workplace.
* The working environment for employees has received a 3.0 rating from several employees.
* Numerous employee have been employed by 0-1 company.
* Most employees had a pay increase of between 0% and 12%.
* The majority of employees have rated their performance with a 3.
#### Bivariant Analysis
* Employee performance ratings do not change based on whether the business analyst, technical lead, data scientist, or developer performs well. 
* Employees with a married status who work overtime are ranked higher than those without a married status.
* Male employees receive higher ratings than female employees; this could be because there are more males than females working for the organisation.
* Employee with backgrounds in medicine and life sciences are given higher ratings.
* Employees who travel infrequently are evaluated higher than those who travel regularly or never.
* Employee performance declines as their total work experience exceeds ten years; nevertheless, performance increases when their salary is boosted by more than 19%. As a result, the company's pay increase has an impact on employee performance. Employees that have seen a larger percentage of pay increases have done well.

### Data Preprocessing
* Checking Null Values- There were no null values
* Categorical Numerical data conversion:- Handle categorical data to Numerical using Label encoding and Mannual encoding.
* Outlier Handling:- Some features such as "NumCompaniesWorked ,TotalWorkExperienceInYears,TrainingTimesLastYear ,ExperienceYearsAtThisCompany,ExperienceYearsInCurrentRole and YearsWithCurrManager"  contains outliers so we are imputing these outliers with the help of IQR. 

### Feature Selection
* There are lot of features in our dataset, to extract important features which really contribute to target feature we need to find corelation coefficient. Corelation coefficient is used to find out the important features in dataset and these features are then used to train the model.
* Drop unique and constant feature: Dropping employee number because this is a constant column 
* Checking Correlation: Checking correlation with the help of heat map, and get the their is no highly correlated feature is present. Heatmap: A heatmap is a graphical representation of data that uses a system of color-coding to represent different values.

### Machine Learning Model 
#### The machine learning models used in this project are

* Support Vector Machine
* Random Forest classifier
 Both machine learning algorithms are best for classification and labelled data. The train and test data are divided and fitted into the model and passed through the machine learning. Since we have already noted the severe imbalance in the values within the target variable, we implement the SMOTE method. The predicted data and test data achieved the accuracy rate of,

* Support Vector machine: 69%
* Artificial Nueral Network: 84%
* Random Forest classifier: 95.88%

### Goal 1. Department wise performances
 In department wise performance, we have to analyze the data from each department present in the category. The data frame has to be separated or sliced according to department wise. In Employee department feature there are six departments available. The performance analysis by the department as follows:

* Sales, Development, and Research & Development departments have a higher number of employees with performance rating 3, indicating average performance is common in these areas.
* Development and Research & Development departments also have a significant number of employees with a performance rating of 4, suggesting these departments are performing well overall.
* Human Resources, Data Science, and Finance departments have a smaller number of high performers (rating 4) and lower performers (rating 2), with most employees falling into the average performance category (rating 3).
* The Development department stands out with a high number of high-performing employees, indicating strong performance within this department.
* Sales and Research & Development departments show a higher variability in performance ratings, indicating a wider range of employee performance.

* The increase in pay will encourage workers to achieve strong financial and psychological outcomes.
