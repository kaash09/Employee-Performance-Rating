### Bivariate Analysis
Checking relationship of variables related to target variable

```python
data.head()
```

```python
data.describe(include='O').T  #Checking data for cat-col.
```

```python
data2=data[['Gender', 'EducationBackground', 'MaritalStatus',
       'EmpDepartment', 'EmpJobRole', 'BusinessTravelFrequency', 'OverTime',
       'Attrition']]
```

```python
data2
```

```python
# plotting how every categorical feature correlated with the target 
plt.figure(figsize=(30,40),facecolor='white')
pltno=1 
for column in data2: #for loop to acess columns form data2
    if pltno <=9:
            
        ax=plt.subplot (5,2,pltno)
        sns.countplot(x=data2[column].dropna(axis=0),hue=data.PerformanceRating)
        plt.xlabel(column,fontsize=30, )
        plt.ylabel('PerformanceRating',fontsize=30)
    pltno+=1
plt.tight_layout()
```

### Insights from bivariate analysis
* Across both Gender and EducationBackground, PerformanceRating 3 is the most frequent.
* The consistency in the distribution of performance ratings across both genders suggests there is no significant bias or difference based on gender.
* The variation in distributions among different education backgrounds indicates that educational background might influence performance ratings to some extent, with some fields showing a higher concentration of mid-range performance ratings
* Married employees have the highest average performance rating, followed by single employees, then divorced employees, and then widowed employees.
* Sales department has the lowest performance rating out of all department followed by R&D.  
* Employees who go on a business travel rarely have a better performancing rating (4).
* Performance rating of 3 is the most common in both the OverTime and Attrition categories, indicating that most employees, regardless of these factors, receive a middle-tier performance rating.
