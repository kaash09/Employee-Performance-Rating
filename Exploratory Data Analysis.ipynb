### EDA
#### Checking the distribution of continuous feature

```python
# Get the continuous features
data.select_dtypes('int').head()
```

```python
# Make a list of continuous feature in nature
cont = data[['DistanceFromHome','EmpHourlyRate','EmpLastSalaryHikePercent','TotalWorkExperienceInYears',
                   'ExperienceYearsAtThisCompany','ExperienceYearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]

plt.figure(figsize=(20,15)) # defining canvas size
plotno = 1 # counter

for column in cont: # iteration of columns / acessing the columns from count 
    if plotno<=8:    # set the limit
        plt.subplot(3,3,plotno) # # plotting 7 graphs (3-rows,3-columns) ,plotnumber is for count
        sns.distplot(x=cont[column]) # Plotting dist plots 
        plt.xlabel(column,fontsize=20)  # assigning name to x-axis and font size is 20
    plotno+=1 # counter increment
plt.tight_layout()
plt.show() # used to hide the storage loction
```

#### CHECK THE SKEWNESS
Skewness:Skewness is a measure of the asymmetry of a distribution. A distribution is asymmetrical when its left and right side are not mirror images. A distribution can have right (or positive), left (or negative), or zero skewness

```python

print('Distance From Home Feature Skewness:',data.DistanceFromHome.skew())
print('Employee Hourly Rate Feature Skewness:',data.EmpHourlyRate.skew())
print('Employee Last Salary Hike Percent Feature Skewness:',data.EmpLastSalaryHikePercent.skew())
print('Total Work Experiance In Year Feature Skewness:',data.TotalWorkExperienceInYears.skew())
print('Experiance Year At This Company Feature Skewness:',data.ExperienceYearsAtThisCompany.skew())
print('Experiance Year In Current Role Feature Skewness:',data.ExperienceYearsInCurrentRole.skew())
print('Year Since Last Promotion Feature Skewness:',data.YearsSinceLastPromotion.skew())
print('Years With Current Manager Feature Skewness:',data.YearsWithCurrManager.skew())
```

### CHECK THE KURTOSIS
Kurtosis: kurtosis is a measure of the "tailedness" of the probability distribution of a real-valued random variable.

```python
print('Distance From Home Feature kurtosis:',data.DistanceFromHome.kurtosis())
print('Employee Hourly Rate Feature kurtosis:',data.EmpHourlyRate.kurtosis())
print('Employee Last Salary Hike Percent Feature kurtosis:',data.EmpLastSalaryHikePercent.kurtosis())
print('Total Work Experiance In Year Feature kurtosis:',data.TotalWorkExperienceInYears.kurtosis())
print('Experiance Year At This Company Feature kurtosis:',data.ExperienceYearsAtThisCompany.kurtosis())
print('Experiance Year In Current Role Feature kurtosis:',data.ExperienceYearsInCurrentRole.kurtosis())
print('Year Since Last Promotion Feature kurtosis:',data.YearsSinceLastPromotion.kurtosis())
print('Years With Current Manager Feature kurtosis:',data.YearsWithCurrManager.kurtosis())
```
